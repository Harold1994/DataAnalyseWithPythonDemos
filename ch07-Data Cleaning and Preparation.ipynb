```python
mnames = ['movie_id','title','genres']
```

```python
all_genres = []
```

```python
for x in movies.genres:
    all_genres.extend(x.split('|'))
```

```python
generes = pd.unique(all_genres)
```

```python
generes
```

```python
zero_matrix = np.zeros((len(movies), len(generes)))
```

```python
np.zeros?
```

```python
dummies = pd.DataFrame(zero_matrix, columns=generes)
```

```python
gen.split("|")
```

```python
dummies
```

```python
pd.DataFrame?
```
