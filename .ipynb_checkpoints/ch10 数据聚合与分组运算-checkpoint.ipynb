```python
#GroupBy机制
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'key1':['a','a','b','b','a'],
                  'key2' : ['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
```

```python
df
```

```python
grouped = df['data1'].groupby(df['key1'])
```

```python
grouped#变量grouped是一个GroupBy对象。它实际上还没有进行任何计算，只是含有一些有关分组键df['key1']的中间数据而已。
```

```python
grouped.mean()
```

```python
mean = df['data1'].groupby([df['key1'],df['key2']]).mean()
```

```python
mean
```

```python
mean.unstack()
```

```python
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
```

```python
years = np.array([2005, 2005, 2006, 2005, 2006])
```

```python
df['data1'].groupby([states, years]).mean()
```

```python
df.groupby('key1').mean()#df['key2']不是数值数据（俗称“麻烦列”），所以被从结果中排除了
```

```python
 df.groupby(['key1', 'key2']).mean()
```

```python
df.groupby(['key1','key2']).size()
```

```python
# 对分组进行迭代
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for (k1, k2), group in df.groupby(['key1', 'key2']):#对于多重键的情况，元组的第一个元素将会是由键值组成的元组：
   ....:     print((k1, k2))
   ....:     print(group)
```

```python
#将这些数据片段做成一个字典
pieces = dict(list(df.groupby('key1')))
```

```python
# 通过字典或Series进行分组
people = pd.DataFrame(np.random.randn(5, 5),
                    columns=['a', 'b', 'c', 'd', 'e'],
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
```

```python
people.iloc[2:3,[1,2]] = np.NaN
```

```python
people
```

```python
mapping = {'a': 'red', 'b': 'red', 'c': 'blue',
   ....:            'd': 'blue', 'e': 'red', 'f' : 'orange'}
```

```python
ny_column = people.groupby(mapping,axis=1)
```

```python
ny_column.sum()
```

```python
map_Series = pd.Series(mapping)
```

```python
map_Series
```

```python
people.groupby(map_Series, axis=1).count()
```

```python
# 通过函数进行分组
# 任何被当做分组键的函数都会在各个索引值上被调用一次，其返回值就会被用作分组名称。
people.index
```

```python
people.groupby(len).sum()
```

```python
# 将函数跟数组、列表、字典、Series混合使用,任何东西在内部都会被转换为数组：
key_list = ['one', 'one', 'one', 'two', 'two']
people.groupby([len, key_list]).min()
```

```python
##数据聚合
# quantile可以计算Series或DataFrame列的样本分位数。
df
```

```python
grouped = df.groupby('key1')
```

```python
# 使用你自己的聚合函数，只需将其传入aggregate或agg方法即可
def peak_to_peak(arr):
    return arr.max() - arr.min()
grouped.agg(peak_to_peak)
```

```python
# 面向列的多函数应用
tips = pd.read_csv('../pydata-book/examples/tips.csv')
```

```python
tips['tip_pct'] = tips['tip']/tips['total_bill']
```

```python
tips.head()
```

```python
grouped =tips.groupby(['day','smoker'])
```

```python
grouped_pct = grouped['tip_pct']
```

```python
grouped_pct.agg("mean")
```

```python
grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
# 如果传入的是一个由(name,function)元组组成的列表，则各元组的第一个元素就会被用作DataFrame的列名
grouped_pct.agg([('foo','mean'),('bar',np.std)])
```

```python
functions = ['count', 'mean', 'max']
```

```python
result = grouped['tip_pct', 'total_bill'].agg(functions)
```

```python
result
```

```python
ftuples = [('Durchschnitt', 'mean'),('Abweichung', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
# 想要对一个列或不同的列应用不同的函数。具体的办法是向agg传入一个从列名映射到函数的字典
grouped.agg({'tip':np.max, 'size':'sum'})
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
   ....:              'size' : 'sum'})
```

```python
# apply：一般性的“拆分－应用－合并”
def top(df, n=5,column='tip_pct'):
    return df.sort_values(by=column)[-5:]
```

```python
top(tips, n = 6)
```

```python
tips.groupby('smoker').apply(top)#top函数在DataFrame的各个片段上调用，然后结果由pandas.concat组装到一起，并以分组名称进行了标记。
#是，最终结果就有了一个层次化索引，其内层索引值来自原DataFrame。
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
```

```python
result
```

```python
result.unstack()
```

```python
# 禁止分组键
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
# 分位数和桶分析
frame = pd.DataFrame({'data1': np.random.randn(1000),'data2': np.random.randn(1000)})
```

```python
quartiles = pd.cut(frame.data1, 4)
```

```python
quartiles[:10]
```

```python
quartiles.value_counts()
```

```python
def get_status (group):
    return {'min':group.min(),'max':group.max,'count':group.count(),'mean' : group.mean()}
grouped = frame.data2.groupby(quartiles)
grouped.apply(get_status).unstack()
```

```python
grouped.describe()
```

```python
grouping = pd.qcut(frame.data1, 10, labels=False)
```

```python
grouped = frame.data2.groupby(grouping)
```

```python
grouped.apply(get_status).unstack()
```

```python
# 随机采样和排列
suits = ['H','S','C','D'] 
card_val=(list(range(1,11)) + [10]*3) *4
```

```python
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
```

```python
base_names
```

```python
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)
```

```python
deck = pd.Series(card_val, index=cards)
```

```python
deck[:13]
```

```python
def draw(deck, n=5):
    return deck.sample(n)
draw(deck)
```

```python
get_suit = lambda card: card[-1]
```

```python
deck.groupby(get_suit).apply(draw,2)#从每种花色中随机抽取两张牌
```

```python
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

```python
# 示例：分组加权平均数和相关系数
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a',                                'b', 'b', 'b', 'b'],
                 'data': np.random.randn(8),
                   'weights': np.random.rand(8)})
```

```python
close_px = pd.read_csv('../pydata-book/examples/stock_px_2.csv', parse_dates=True, index_col= 0)
```

```python
close_px.info()
```

```python
close_px.head()
```

```python
spx_corr = lambda x: x.corrwith(x['SPX'])
```

```python
rets = close_px.pct_change().dropna()
```

```python
get_year = lambda x: x.year
```

```python
by_year= rets.groupby(get_year)
```

```python
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))
```

```python
# 示例：组级别的线性回归
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['Intercept'] = 1.
    result = sm.OLS(X,Y).fit()
    return result.params
```

```python
by_year.apply(regress,'AAPL',['SPX'])
```
